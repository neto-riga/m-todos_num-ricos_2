# Ejercicio 13. Diferencias finitas
---

Sea la función:
$$
f(x) = 4xln(x) + \frac{\cos(x-3)}{\sqrt{1+x^2}}
$$

* Estimar la derivada para $x=1.5$ con $h=0.1$, $0.05$ y $0.025$ empleando la fórmula de diferencia centrada.
* Tabular la función en el intervalo $[0.5,2.3]$ con intervalor de $h=0.1$ y estimar la derivada en cada punto, para $[0.6,2.2]$.
* Para los incisos anteriores comparar con la solución analítica.


```python
import numpy as np 
import pandas as pd 
import math
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

Declaramos la función para llenar listas.

```python
def Listas (val,h,n):
  f = lambda x: (4*x*np.log(x)) + ((np.cos(x-3))/(np.sqrt(1 + x**2)))
  x = [val - h]
  fx = [f(x[0])]
  for i in range(n):
    x.append(x[i]+h)
    i+=1
    fx.append(f(x[i]))
  return x,fx
```

Definimos las diferencias centradas.

```python
def Dife_cen(x,fx,h):
  dif = []
  for i in range(len(x)-2):
    dif.append((1/(2*h)) * (fx[i+2]-fx[i]))
  return dif
```

Tabulamos para las diferentes $h$ 

```python
h = [0.1,0.05,0.025]
val = 1.5
x1,fx1 = Listas(val,h[0],2)
tabla1 = {'x': x1, 'f(x)': fx1}
print(f'Para h={h[0]}: \n')
df1 = pd.DataFrame(tabla1)
display(df1)
x2,fx2 = Listas(val,h[1],2)
tabla2 = {'x': x2, 'f(x)': fx2}
print(f'Para h={h[1]}: \n')
df2 = pd.DataFrame(tabla2)
display(df2)
x3,fx3 = Listas(val,h[2],2)
tabla3 = {'x': x3, 'f(x)': fx3}
print(f'Para h={h[2]}: \n')
df3 = pd.DataFrame(tabla3)
display(df3)
```

Declaramos $f'(x)$

```python
def D(x):
  r = (4*np.log(x)) + 4 
  z = (((1+x**2)*(-np.sin(x-3))) - (x*np.cos(x-3))) / ((1+x**2)**(3/2))
  return r+z
```

Aproximamos a la derivadas.

```python
Sol=[]
Sol.append(Dife_cen(x1,fx1,h[0]))
Sol.append(Dife_cen(x2,fx2,h[1]))
Sol.append(Dife_cen(x3,fx3,h[2]))
Err=[]
for i in range(len(Sol)):
  Err.append(abs(D(val) - Sol[i]))
tabla4 = {'h': h,
          'Dif. centrada': Sol,
          'Error absoluto': Err
}
df4 = pd.DataFrame(tabla4)
df4
```

Tabulamos la función en el intervalo $[0.5,2.3]$ con $h=0.1$

```python
x4,fx4 = Listas(0.6,0.1,18)
tabla5 = {'x': x4, 'f(x)': fx4}
df5 = pd.DataFrame(tabla5)
df5
```

Apriximamos la derivada y obtenemos el error absoluto del intervalo $[0.6,2.2]$

```python
lis = Dife_cen(x4,fx4,0.1)
Err2=[]
for i in range(len(lis)):
  Err2.append(abs(D(x4[1:-1][i]) - lis[i]))

tabla6={'x': x4[1:-1], 'Dif. Centrada': lis, 'Error abs': Err2}
df6 = pd.DataFrame(tabla6)
df6
```
